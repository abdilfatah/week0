```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("benin-malanville.csv")
```

```python
pairs = [
    ('GHI', 'Tamb'),
    ('WS', 'WSgust'),
]
```

```python
plt.figure(figsize=(10, 5))
```

```python
for i, (x_var, y_var) in enumerate(pairs):
    plt.subplot(1, len(pairs), i+1)
    plt.scatter(df[x_var], df[y_var], alpha=0.5)
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(f'{x_var} vs. {y_var}')
```

```python
plt.tight_layout()
plt.show()
```

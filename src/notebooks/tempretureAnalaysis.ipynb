```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("benin-malanville.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.scatter(df['Tamb'], df['TModA'], label='TModA', alpha=0.5)
plt.scatter(df['Tamb'], df['TModB'], label='TModB', alpha=0.5)
plt.xlabel('Ambient Temperature (Tamb)')
plt.ylabel('Module Temperature')
plt.title('Module Temperatures vs. Ambient Temperature')
plt.legend()
plt.show()
```

```python
corr_TModA_Tamb = df['TModA'].corr(df['Tamb'])
corr_TModB_Tamb = df['TModB'].corr(df['Tamb'])
print(f"Correlation TModA-Tamb: {corr_TModA_Tamb:.2f}")
print(f"Correlation TModB-Tamb: {corr_TModB_Tamb:.2f}")
```
